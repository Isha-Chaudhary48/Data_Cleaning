```python
# What is the distribution of employee age
```

```python
plt.hist(df["Age"],bins=5)
plt.xlabel("Age")
```

```python
# What is the average salary by city?
avg_city=df.groupby('City')['Salary'].mean()
plt.bar(avg_city.index,avg_city.values,color="skyblue")
plt.xlabel("Cities")
plt.ylabel("Average salary")
plt.title("average salary by city")
```

```python
# How does salary vary across different cities?
plt.figure(figsize=(15,6))
df.boxplot('Salary',by="City",color="orange", medianprops=dict(color='blue'), whiskerprops=dict(color='green'),capprops=dict(color='skyblue'))
plt.ylabel("Salary")
plt.show()
```

```python
# What is the relationship between Age and Salary?
plt.figure(figsize=(5,10))
plt.scatter(df['Age'],df['Salary'],color="lightgreen")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# How many employees live in each city?
employee_live=df.groupby('City')['City'].count()
plt.bar(employee_live.index,employee_live.values)
plt.xlabel("City")
plt.ylabel("Number of Employees")
plt.title("Number of Employees in Each City")

plt.show()
```

```python
# What is the salary distribution in the company?
plt.hist(df['Salary'])
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.title("Salary Distribution")
plt.show()
```

```python
# How many employees joined the company each year?
plt.plot(df["Joining_date"].dt.year)

plt.show()
```

```python
# What is the correlation between Age and Salary?
df["Age"].corr(df['Salary'])
```

```python
plt.scatter(df["Age"],df['Salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation between Age and Salary")
plt.show()
```

```python
import seaborn as sns
corr = df[['Age','Salary']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
# Which employees have the highest salaries
employeeSalary= df.groupby('Customer_id')['Salary'].max()
employee=employeeSalary.sort_values(ascending=False).head(5)
plt.bar(employee.index,employee.values,linewidth=10)
plt.xlabel("Customer ID")
plt.ylabel("Salary")
plt.title("Highest Salary by Customer")
plt.show()
```
