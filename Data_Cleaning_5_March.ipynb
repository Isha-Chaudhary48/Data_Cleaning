```python
# histogram : this chart is used to show the distribution of numerical data(continous no's)
```

```python
# Salary Distribution of Customers
# bins are used to decide how many bars (ranges) the histogram will have
# here salary is divided into 10 salary ranges
plt.hist(df["Salary"],bins=10,color="green")
plt.xlabel("Salary")
plt.ylabel("No of customers")
plt.show()
```

```python
# age distribution of customer
plt.hist(df["Age"],bins=15)
plt.xlabel("Age")
plt.ylabel("No of customers")
plt.show()
```

```python
# scatter plot : used to show relationship b/w two numerical variables
```

```python
# Age vs Salary Relationship
plt.figure(figsize=(8,6))
plt.scatter(df["Age"],df["Salary"],alpha=0.6)
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# boxplot : used to visualize distribution of numerical data and to detect outliers
# it contains quartiles min,max median, outliers
```

```python
# Salary Spread and Outliers
plt.boxplot(df["Salary"])
plt.title("Salary distribution")
plt.show()
```

```python
df["Age_Group"] = pd.cut(df["Age"], bins=[19,30,40,50,60])

df.boxplot(column="Salary", by="Age_Group")
plt.show()
```

```python

df
```

```python

groups = df.groupby("Age_Group")["Salary"].apply(list)
groups
```

```python
plt.boxplot(groups,tick_labels=groups.index)
plt.show()
```

```python
# Customer Age Groups
# we have Age_groups 
age_count = df['Age_Group'].value_counts()
age_count
```

```python
plt.bar(age_count.index.astype(str),age_count.values)
plt.title("Customer age groups")
plt.xlabel("Age groups")
plt.ylabel("Number of customers")
plt.show()
```

```python
# Average Salary by Age Group
average_salary=df.groupby('Age_Group')["Salary"].mean()
average_salary
```

```python
plt.bar(average_salary.index.astype(str),average_salary.values)
plt.xlabel("Age Group")
plt.ylabel("Average Salary")
plt.title("Average Salary by Age Group")
plt.show()
```

```python
# line chart is used to show how trends changes over time
# use line chart when you want to show continous data
```

```python
# What is the average salary across different departments?
average_salary=df.groupby(df['Department'])['Salary'].mean()
average_salary
```

```python
plt.figure(figsize=(10,5))
plt.bar(average_salary.index,average_salary.values,color='skyblue')
plt.xlabel("Departments")
plt.ylabel("Average Salary")
plt.title("average salary across different departments")
plt.show()
```

```python
# How does salary vary among different departments?
plt.figure(figsize=(15,6))
df.boxplot(column="Salary" , by="Department")
plt.suptitle("")   # removes extra title
plt.xlabel("Department")
plt.ylabel("Salary")

plt.show()
```
