```python
def missingCity():
    inputSerialNo= int(input("Enter serial no"))
    city = input("Enter city")
    mask = ((missing_cities['Serial_no'] == inputSerialNo) & (missing_cities['City'].isna()))
    if mask.any():
        missing_cities.loc[mask,'City'] = city
        print("City updated successfully")
    else:
        print("City already exists or invalid serial no ")
       
missingCity()
            
    
        
```
